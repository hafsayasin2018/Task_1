```python
Data_set = r"C:\Users\Hafsa\Pandas\pandas_project\Sales_Analyses\SalesAnalysis\Sales_Data"
Sales_files = [file for file in os.listdir(Data_set) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in Sales_files:
    current_data = pd.read_csv(Data_set+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy1.csv", index=False)
```

```python
all_data = pd.read_csv("all_data_copy1.csv")
all_data.head() # return first 5 entites
all_data.tail() # return last 5 entites
```

Clean up data:

```python
# drop NAN values
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

### Get rid of text in column

```python
# Get rid r in order date colmun
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

### Make data correct type:


```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```
